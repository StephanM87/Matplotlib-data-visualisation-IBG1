# Upscaling of the secret reaction from 10 to 50 mL

In this experiment, data was upscaled from 10 to 50 mL. The 10 mL reaction was repeated over all 3 times and the 50 mL reaction one time. Yielding the reactions:

- upscale_10mL_1
- upscale_10mL_2
- upscale_10mL_3
- upscale_50mL_1


# Importing the data sets from the measurements

At first we need to extract the data from the excel file. One challenge: There are multiple _sheets_ in this excel file. This is luckily not a problem, we can use the `sheet_name` attribute of pandas to extract the sheet names:

```
xl = pd.ExcelFile(r"scale-up-10ml.xlsx")
print(xl.sheet_names)
```

Afterwards we can store the data in 4 different data frames:

```
df_upscale_10mL_1 = pd.read_excel(r"scale-up-10ml.xlsx",'Sclae up-10mL')
df_upscale_10mL_2 = pd.read_excel(r"scale-up-10ml.xlsx",'Scale-up-10mL-exp-2')
df_upscale_10mL_3 = pd.read_excel(r"scale-up-10ml.xlsx",'Scale-up-10-mL-exp-3')
df_upscale_50mL_1 = pd.read_excel(r"scale-up-10ml.xlsx",'Scale-up-50-mL-exp1')
```

````python
import pandas as pd


# Print the available sheet names

xl = pd.ExcelFile(r"scale-up-10ml.xlsx")
print(xl.sheet_names)

# Read in the data of the reactions:

df_upscale_10mL_1 = pd.read_excel(r"scale-up-10ml.xlsx",'Sclae up-10mL')
df_upscale_10mL_2 = pd.read_excel(r"scale-up-10ml.xlsx",'Scale-up-10mL-exp-2')
df_upscale_10mL_3 = pd.read_excel(r"scale-up-10ml.xlsx",'Scale-up-10-mL-exp-3')
df_upscale_50mL_1 = pd.read_excel(r"scale-up-10ml.xlsx",'Scale-up-50-mL-exp1')

# print the heads of the reaction to check if everything went valid

print(df_upscale_10mL_1.head())
print(df_upscale_10mL_2.head())
print(df_upscale_10mL_3.head())
print(df_upscale_50mL_1.head())

````

# Create a graph to display the individual reactions

Once we have our data extracted, let's get crazy! The first step is of cource to get an overview over all data so, lets visualise every data set in one diagram:

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(4)
fig.suptitle('upscaling reactions')
axs[0].plot(df_upscale_10mL_1['Sample [min]'], df_upscale_10mL_1['(S)-3-OH-PAC'])
axs[0].plot(df_upscale_10mL_1['Sample [min]'], df_upscale_10mL_1['3-OH-BA'])
axs[1].plot(df_upscale_10mL_2['Sample [min]'], df_upscale_10mL_2['(S)-3-OH-PAC'])
axs[1].plot(df_upscale_10mL_2['Sample [min]'], df_upscale_10mL_2['3-OH-BA'])
axs[2].plot(df_upscale_10mL_3['Sample [min]'], df_upscale_10mL_2['(S)-3-OH-PAC'])
axs[2].plot(df_upscale_10mL_3['Sample [min]'], df_upscale_10mL_2['3-OH-BA'])
axs[3].plot(df_upscale_50mL_1['Sample [min]'], df_upscale_10mL_2['(S)-3-OH-PAC'])
axs[3].plot(df_upscale_50mL_1['Sample [min]'], df_upscale_10mL_2['3-OH-BA'])

# Do you like this look?! I don't! So lets rebuild this!
````
