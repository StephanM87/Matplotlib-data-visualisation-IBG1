# Concentration-Time curve of an immobilised enzyme



In this experiment, a secret immobilised enzyme was used to measure concentration of substrate and products over time. THe Enzyme was immobilised on three different beads:

- Opal
- Amber
- Coral

Two different chemicals were measured:

3OH_BA (3-OH-Benzaldehyde) **(Substrate)**

3OH_PAC (3-OH-PAC) **Product**

The measurements consits of triplicates. 

```python
import pandas as pd


# Read in the data of the reactions:

df_amber = pd.read_excel(r"immobilised_biocatalyst.xlsx", "Amber")
df_amber.head()

# get all the available sheets

xl = pd.ExcelFile(r"immobilised_biocatalyst.xlsx")
print(xl.sheet_names)

df_amber_mean = pd.read_excel(r"immobilised_biocatalyst.xlsx", "Amber u and s")


```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 250

fig, (ax, ax1, ax2) = plt.subplots(1,3)
fig.suptitle('Amber immobilisation')
ax.plot(df_amber['Zeit [h]'], df_amber['3-OH-BA-1'])
ax.plot(df_amber['Zeit [h]'], df_amber['3-OH-PAC-1'])
ax.set_xlabel("Time in h")
ax.set_ylabel("concentration in mmol/L")
ax1.plot(df_amber['Zeit [h]'], df_amber['3-OH-BA-2'])
ax1.plot(df_amber['Zeit [h]'], df_amber['3-OH-PAC-2'])
ax1.set_xlabel("Time in h")
ax2.plot(df_amber['Zeit [h]'], df_amber['3-OH-BA-2.1'])
ax2.plot(df_amber['Zeit [h]'], df_amber['3-OH-PAC-2.1'])
ax2.set_xlabel("Time in h")

# Thats better, but the y-axis is so streched, right and the images so small right?!
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = df_amber_mean

# show me all the columns
print(df.columns)

mean_3OH_BA = df["3-OH-BA"]
mean_3OH_PAC = df["3-OH-PAC"]
error_3OH_BA = df["stabw 3-OH-BA "]
error_3OH_PAC = df["stabw 3-OH-PAC "]



# this enlarges the canvas

plt.rcParams['figure.dpi'] = 200

fig, ax = plt.subplots()

ax.errorbar(df['Zeit [h]'], mean_3OH_BA, label="3-OH-BA-mean", linestyle="", color="black", marker="o", ms=3.5, yerr=error_3OH_BA, capsize=3)
ax.errorbar(df['Zeit [h]'], mean_3OH_PAC, label="3-OH-PAC-mean", linestyle="", color="grey",  marker="s", yerr=error_3OH_PAC, capsize=3)

ax.set_xlabel("Time in h")
ax.set_ylabel("Concentration in mmol/L")
ax.set_title("Time conentration curve of secret reaction")
ax.axvline(7, color="black", label="equilibrium reached", linestyle="--")

ax.legend(bbox_to_anchor=(1.4, 0.5))

plt.show()
```
