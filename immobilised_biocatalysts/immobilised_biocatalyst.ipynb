# Concentration-Time curve of an immobilised enzyme



In this experiment, a secret immobilised enzyme was used to measure concentration of substrate and products over time. 

The enzyme catalyses the secret reaction:

2 3OH_BA --> 1 3OH_PAC

To perform the experiment, the enzyme was immobilised to three different beads:

- Opal
- Amber
- Coral

Two different chemicals were measured:

3OH_BA **(Substrate)**

3OH_PAC **Product**

The measurements consits of triplicates. 

More specs will follow

```python
import pandas as pd


# Read in the data of the reactions:

df_amber = pd.read_excel(r"immobilised_biocatalyst.xlsx", "Amber")
df_amber.head()


```

# Create a graph to display the individual reactions

Once we have our data extracted, let's get crazy! The first step is of cource to get an overview over all data so, lets visualise every data set in one diagram:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")

plt.rcParams['figure.dpi'] = 400

fig, (ax, ax1, ax2) = plt.subplots(1,3)
fig.suptitle('Amber immobilisation')
ax.plot(df_amber['Zeit [h]'], df_amber['3-OH-BA-1'], color="black")
ax.plot(df_amber['Zeit [h]'], df_amber['3-OH-PAC-1'], color="orange")
ax.set_ylabel("concentration in mmol/L")
ax.set_xlabel("time in h")
ax1.plot(df_amber['Zeit [h]'], df_amber['3-OH-BA-2'], color="black")
ax1.plot(df_amber['Zeit [h]'], df_amber['3-OH-PAC-2'], color="orange")
ax2.plot(df_amber['Zeit [h]'], df_amber['3-OH-BA-2.1'], color="black")
ax1.set_xlabel("time in h")
ax2.plot(df_amber['Zeit [h]'], df_amber['3-OH-PAC-2.1'], color="orange")
ax2.set_xlabel("time in h")

# Thats better, but the y-axis is so streched, right and the images so small right?!
```
