# How to visualise your data using matplotlib

To realise your data visualisation, you will need 3 different practises:

1. Your Data
Of course you need data. Since everyone in this world is using excel, there is a tool called pandas, which will allow you to extract data from a excel file



```python
# First import pandas

import pandas as pd

# Import the data from the excel file

df = pd.read_excel(r"scale-up-10ml.xlsx")
print(df.head())

'''

Now, to get access to the individual data in the colums you need to extract 
them with df["column name"]. If you are a little bit like me, then you will have problems here,
bexause you misspell something or have a copy pase error. To simplyfy this I propose the following
workflow:

1. Define the vairables in which you want to store your data in:

time = None
PAC = None
BA = None

2. extract the column names with df.columns

columns = df.columns

3. Connect the values with your variables

Last but not least connect your variables with the column names:

'''


time = None
PAC = None
BA = None


columns = df.columns

print(columns)

time = df[columns[0]]
PAC = df[columns[1]]
BA = df[columns[2]]


```

```python
# First visualisation with Matplitlib

# Now you have your data available and stored in the variables above. Time to do some plotting

# First you will need to import matplotlib

import matplotlib.pyplot as plt

fig, ax  = plt.subplots()

ax.plot(time, PAC, label="concentration of PAC over time", color="red", marker="d")

plt.show()

# Now you have created your first plot, lets add in the next cell, the second column with BA


```

```python
# Since you have imported matplotlib in the cell above, you won't have to do it here again

# Wait, didn't we use this command above? Yes, but it doesn't matter, the variable here will simply be overridden

fig, ax = plt.subplots()

ax.plot(time, PAC, label="concentration of PAC over time", color="red", marker="d")
ax.plot(time, BA, label="concentration of PAC over time", color="red", marker="o")
plt.show()

# Oh fuck! Thats bad they have very different ranges, how you will solve this, see in the cell below
```

```python
# To solve the issue above, we need a mehtod that both curves will be visualised in two different ranges:

fig, ax = plt.subplots()

# the ax.twinx tells our program that both new y axes share the same x-axis

ax2 = ax.twinx()

ax.plot(time, PAC, label="concentration of PAC over time", color="red", marker="d")
ax2.plot(time, BA, label="concentration of PAC over time", color="blue", marker="o")
plt.show()

# Now we have both graphs in a nice range. that the BA curve looks like shit is not the fault of matlotlib
```

```python
# Maybe you have already seen that we have the curves, but the description is still missing right?

# Let's do this:

# First, copy the code from above

fig, ax = plt.subplots()

# the ax.twinx tells our program that both new y axes share the same x-axis

ax2 = ax.twinx()

ax.plot(time, PAC, label=" PAC over time", color="red", marker="d")
ax2.plot(time, BA, label="BA over time", color="blue", marker="o")


# Now lets add an diagram title

plt.title("time dependent concentration curve")

# And the title of the x-axis

ax.set_xlabel("time in minutes")

# The title of the y-axisess

ax.set_ylabel("concentration in mmol/L")
ax2.set_ylabel("concentrationin mmol/L")


plt.show()
```
